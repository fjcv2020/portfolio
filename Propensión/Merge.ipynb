```python
import pandas as pd
```

```python
results_pension = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDS - Grupo 7/Capstone/Propensión/results_pension_plan.csv')
results_debit = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDS - Grupo 7/Capstone/Propensión/results_debit.csv')
results_ema = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDS - Grupo 7/Capstone/Propensión/results_ema.csv')
results_payroll = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDS - Grupo 7/Capstone/Propensión/results_payroll.csv')
results_emc = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MDS - Grupo 7/Capstone/Propensión/results_emc.csv')
```

```python
# Primero, unimos los dos primeros dataframes
df_merged = pd.merge(results_pension, results_debit, on='pk_cid', how='outer')

# Luego unimos los siguientes dataframes, uno por uno
df_merged = pd.merge(df_merged, results_payroll, on='pk_cid', how='outer')
df_merged = pd.merge(df_merged, results_ema, on='pk_cid', how='outer')
df_merged = pd.merge(df_merged, results_emc, on='pk_cid', how='outer')
```

```python
# Verifica si hay duplicados basado en la columna 'pk_cid'
duplicates = df_merged.duplicated(subset='pk_cid')

# Imprime el número de duplicados
print(f"Number of duplicate rows based on 'pk_cid': {duplicates.sum()}")

# Si quieres ver los duplicados, puedes hacerlo así:
print(df_merged[df_merged.duplicated(subset='pk_cid', keep=False)].sort_values(by='pk_cid'))
```
